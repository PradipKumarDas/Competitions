```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
players_from_hackathon = pd.read_csv("playersList.csv")
```

```python
players_from_hackathon = pd.DataFrame(
    players_from_hackathon.Players.apply(lambda player: 
                                         " ".join(player.split(" ")[-1:]+player.split(" ")[:-1])))
```

```python
players_from_hackathon.Players = players_from_hackathon.Players.str.lower()
```

```python
players_from_hackathon = pd.DataFrame(players_from_hackathon.Players.sort_values(ignore_index=True))
```

```python
players_from_hackathon
```

```python
players_from_cricsheet = pd.read_csv("Data/all_matches.csv")[["striker", "non_striker", "bowler"]]
```

```python
players_from_cricsheet = pd.DataFrame(
    players_from_cricsheet.striker.append(
        players_from_cricsheet.non_striker).append(
            players_from_cricsheet.bowler).unique(), 
    columns=["Players"])
```

```python
players_from_cricsheet = pd.DataFrame(
    players_from_cricsheet.Players.apply(lambda player: 
                                         " ".join(player.split(" ")[-1:]+player.split(" ")[:-1])))
```

```python
players_from_cricsheet.Players = players_from_cricsheet.Players.str.lower()
```

```python
players_from_cricsheet = pd.DataFrame(players_from_cricsheet.Players.sort_values(ignore_index=True))
```

```python
players_from_cricsheet
```

```python
for player in players_from_cricsheet.Players:
    print(player)
```

```python
players_from_cricsheet.Players.isin(players_from_hackathon.Players).value_counts()
```

```python
players_from_hackathon.Players.isin(players_from_cricsheet.Players).value_counts()
```
